```python
# Task 2 – Automated Report Generation

import pandas as pd
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet

# Step 1: Read data
data = pd.read_csv("data.csv")

# Step 2: Perform analysis
average_marks = data["Marks"].mean()
highest_marks = data["Marks"].max()
lowest_marks = data["Marks"].min()

# Step 3: Generate PDF
doc = SimpleDocTemplate("report.pdf", pagesize=A4)
styles = getSampleStyleSheet()
elements = []

elements.append(Paragraph("Student Performance Report", styles['Title']))
elements.append(Spacer(1, 20))

elements.append(Paragraph(f"Total Students: {len(data)}", styles['BodyText']))
elements.append(Paragraph(f"Average Marks: {average_marks:.2f}", styles['BodyText']))
elements.append(Paragraph(f"Highest Marks: {highest_marks}", styles['BodyText']))
elements.append(Paragraph(f"Lowest Marks: {lowest_marks}", styles['BodyText']))

doc.build(elements)

print("PDF Report Generated Successfully!")
```
